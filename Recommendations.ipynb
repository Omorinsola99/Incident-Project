Summary
This project analyzed incidents of accidents, violence, and conflict across Nigeria using a structured dataset. 

The data was first cleaned to handle missing values, remove duplicates, and correct inconsistencies in state names and formatting. After preparation, exploratory data analysis (EDA) was conducted to identify patterns, trends, and relationships within the dataset.

Various visualizations were used to better understand the distribution of incidents, the most affected states, and the number of deaths associated with each incident. Time-based analysis also helped reveal how incidents have changed over the years.

Overall, the analysis provided insights into the frequency, location, and severity of incidents across Nigeria.

Conclusion
The analysis revealed that certain states experience a significantly higher number of incidents compared to others, indicating regional disparities in security and safety conditions. Additionally, while most incidents result in relatively low casualties, a small number of events account for a large proportion of deaths.

There is also a noticeable relationship between the frequency of incidents and the number of deaths, suggesting that areas with more frequent incidents tend to experience higher overall casualties.

Over time, the number of incidents has shown fluctuations, which may reflect changes in security measures, reporting systems, or socio-political conditions.

These findings highlight the need for targeted interventions, improved security strategies, and better emergency response systems in high-risk areas to reduce both the frequency and impact of such incidents.
